```python
import numpy as np
import pandas as pd

from scipy import stats
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv('/Users/siddharth/Desktop/btp/BTP/week2/data.csv')
```

```python
data.head(10)
```

```python
data.columns
```

```python
#we drop columns which are not needed

data.drop('id', axis=1, inplace=True)
data.drop('Unnamed: 32', axis=1, inplace=True)
```

```python
# since we have only one full data set, we just split it into 2 parts for test and train

## by trying different 'split' values we get different accuary, low split values 0.1,0.2 give >92% accuaracy but
## that is on this data set which we don't want. Hence, we take 0.4 as a good general value

train, test = train_test_split(data, test_size = 0.4)
```

```python
features = [
    'diagnosis',
    'perimeter_se',
    'perimeter_worst',
    'concavity_worst',
    'concave points_worst',
    'texture_worst',
    'smoothness_worst',
    'symmetry_worst'
]
```

```python
train = train[train['perimeter_se'] < 15]
```

```python
#we put select features into predictors
predictors = features[2:]

train_X = train[predictors]
train_y = train.diagnosis

test_X = test[predictors]
test_y = test.diagnosis
```

```python
print predictors
```

```python
model = LogisticRegression()
model.fit(train_X, train_y)
```

```python
predictions = model.predict(test[predictors])
accuracy = metrics.accuracy_score(predictions, test['diagnosis'])
print("Accuracy : %s" % "{0:.2%}".format(accuracy))
```
